```python
# ============================================================
# Conditional GAN Implementation for Person Face Sketches
# ============================================================

# ================================
# 1. Environment Setup
# ================================

# Import necessary libraries
import os
import numpy as np
import matplotlib.pyplot as plt
import shutil

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import torchvision

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Check device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f'Using device: {device}')

# ================================
# 2. Accessing and Preparing the Dataset
# ================================

# Define source directories from the Kaggle input
source_train_sketch_dir = '/kaggle/input/person-face-sketches/train/sketches/'
source_train_photo_dir = '/kaggle/input/person-face-sketches/train/photos/'

# Define destination directories in the working directory
destination_train_sketch_dir = '/kaggle/working/data/train/sketches/'
destination_train_photo_dir = '/kaggle/working/data/train/photos/'

# Create destination directories if they don't exist
os.makedirs(destination_train_sketch_dir, exist_ok=True)
os.makedirs(destination_train_photo_dir, exist_ok=True)

# Function to copy images from source to destination
def copy_images(source_dir, dest_dir):
    """
    Copies images from source_dir to dest_dir.
    Only copies files with .png, .jpg, or .jpeg extensions.
    """
    for filename in os.listdir(source_dir):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            shutil.copy(os.path.join(source_dir, filename), dest_dir)

# Copy training sketches and photos
copy_images(source_train_sketch_dir, destination_train_sketch_dir)
copy_images(source_train_photo_dir, destination_train_photo_dir)

print("Training dataset organized into 'data/train/sketches/' and 'data/train/photos/'.")

# Verify the number of images copied
num_train_sketches = len([f for f in os.listdir(destination_train_sketch_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
num_train_photos = len([f for f in os.listdir(destination_train_photo_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
print(f"Number of training sketches: {num_train_sketches}")
print(f"Number of training photos: {num_train_photos}")

# ================================
# 3. Defining the cGAN Architecture
# ================================

# Hyperparameters
batch_size = 64
learning_rate = 0.0002
num_epochs = 200
image_size = 64
num_channels = 3  # RGB images
feature_maps = 64  # Base number of feature maps

# Define the Generator
class Generator(nn.Module):
    def __init__(self, input_channels=3, output_channels=3, feature_maps=64):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            # Input: (input_channels) x 64 x 64
            nn.Conv2d(input_channels, feature_maps, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(feature_maps),
            nn.ReLU(True),

            # State: (feature_maps) x 32 x 32
            nn.Conv2d(feature_maps, feature_maps * 2, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(feature_maps * 2),
            nn.ReLU(True),

            # State: (feature_maps*2) x 16 x 16
            nn.Conv2d(feature_maps * 2, feature_maps * 4, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(feature_maps * 4),
            nn.ReLU(True),

            # State: (feature_maps*4) x 8 x 8
            nn.ConvTranspose2d(feature_maps * 4, feature_maps * 2, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(feature_maps * 2),
            nn.ReLU(True),

            # State: (feature_maps*2) x 16 x 16
            nn.ConvTranspose2d(feature_maps * 2, feature_maps, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(feature_maps),
            nn.ReLU(True),

            # State: (feature_maps) x 32 x 32
            nn.ConvTranspose2d(feature_maps, output_channels, kernel_size=4, stride=2, padding=1, bias=False),
            nn.Tanh()
            # Output: (output_channels) x 64 x 64
        )

    def forward(self, x):
        return self.main(x)

# Define the Discriminator
class Discriminator(nn.Module):
    def __init__(self, input_channels=3, output_channels=3, feature_maps=64):
        super(Discriminator, self).__init__()
        # Concatenate sketch and image channels
        self.main = nn.Sequential(
            # Input: (input_channels * 2) x 64 x 64
            nn.Conv2d(input_channels * 2, feature_maps, kernel_size=4, stride=2, padding=1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),

            # State: (feature_maps) x 32 x 32
            nn.Conv2d(feature_maps, feature_maps * 2, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(feature_maps * 2),
            nn.LeakyReLU(0.2, inplace=True),

            # State: (feature_maps*2) x 16 x 16
            nn.Conv2d(feature_maps * 2, feature_maps * 4, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(feature_maps * 4),
            nn.LeakyReLU(0.2, inplace=True),

            # State: (feature_maps*4) x 8 x 8
            nn.Conv2d(feature_maps * 4, feature_maps * 8, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(feature_maps * 8),
            nn.LeakyReLU(0.2, inplace=True),

            # State: (feature_maps*8) x 4 x 4
            nn.Conv2d(feature_maps * 8, 1, kernel_size=4, stride=1, padding=0, bias=False),
            nn.Sigmoid()
            # Output: 1 x 1 x 1
        )

    def forward(self, sketch, img):
        # Concatenate sketch and image along the channel dimension
        x = torch.cat([sketch, img], dim=1)
        return self.main(x).view(-1, 1).squeeze(1)

# Initialize the Generator and Discriminator
def initialize_models():
    netG = Generator(input_channels=num_channels, output_channels=num_channels, feature_maps=feature_maps).to(device)
    netD = Discriminator(input_channels=num_channels, output_channels=num_channels, feature_maps=feature_maps).to(device)
    return netG, netD

netG, netD = initialize_models()

print("Generator Architecture:\n", netG)
print("\nDiscriminator Architecture:\n", netD)

# Initialize Weights
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)

netG.apply(weights_init)
netD.apply(weights_init)

# ================================
# 4. Data Loading and Preprocessing
# ================================

# Custom Dataset for Paired Sketches and Photos
class FaceSketchDataset(Dataset):
    def __init__(self, sketch_dir, photo_dir, transform=None):
        super(FaceSketchDataset, self).__init__()
        self.sketch_dir = sketch_dir
        self.photo_dir = photo_dir
        self.sketch_images = sorted([f for f in os.listdir(sketch_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
        self.photo_images = sorted([f for f in os.listdir(photo_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
        self.transform = transform

        assert len(self.sketch_images) == len(self.photo_images), "Mismatch between sketches and photos"

    def __len__(self):
        return len(self.sketch_images)

    def __getitem__(self, idx):
        sketch_path = os.path.join(self.sketch_dir, self.sketch_images[idx])
        photo_path = os.path.join(self.photo_dir, self.photo_images[idx])

        sketch = Image.open(sketch_path).convert('RGB')  # Ensure 3 channels
        photo = Image.open(photo_path).convert('RGB')

        if self.transform:
            sketch = self.transform(sketch)
            photo = self.transform(photo)

        return sketch, photo

# Define transformations
transform = transforms.Compose([
    transforms.Resize((image_size, image_size)),
    transforms.ToTensor(),
    transforms.Normalize([0.5 for _ in range(num_channels)], [0.5 for _ in range(num_channels)])
])

# Initialize Dataset and DataLoader
sketch_dir = destination_train_sketch_dir
photo_dir = destination_train_photo_dir

# Verify that directories exist
if not os.path.exists(sketch_dir) or not os.path.exists(photo_dir):
    raise FileNotFoundError(f"Sketch directory '{sketch_dir}' or photo directory '{photo_dir}' does not exist.")

dataset = FaceSketchDataset(sketch_dir, photo_dir, transform=transform)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True, num_workers=4)

print(f"Dataset loaded with {len(dataset)} samples.")

# ================================
# 5. Loss Functions and Optimizers
# ================================

# Loss Function
criterion = nn.BCELoss()

# Labels
real_label = 1.
fake_label = 0.

# Optimizers
optimizerD = optim.Adam(netD.parameters(), lr=learning_rate, betas=(0.5, 0.999))
optimizerG = optim.Adam(netG.parameters(), lr=learning_rate, betas=(0.5, 0.999))

# ================================
# 6. Training the cGAN
# ================================

# Lists to keep track of progress
img_list = []
G_losses = []
D_losses = []
iters = 0

print("Starting Training Loop...")
for epoch in range(num_epochs):
    for i, data in enumerate(dataloader, 0):
        ############################
        # (1) Update D network
        ###########################
        netD.zero_grad()
        sketches, real_images = data
        sketches = sketches.to(device)
        real_images = real_images.to(device)
        b_size = real_images.size(0)
        label = torch.full((b_size,), real_label, dtype=torch.float, device=device)

        # Forward pass real batch through D
        output = netD(sketches, real_images)
        errD_real = criterion(output, label)
        errD_real.backward()
        D_x = output.mean().item()

        # Generate fake images
        fake_images = netG(sketches)
        label.fill_(fake_label)

        # Classify all fake batch with D
        output = netD(sketches, fake_images.detach())
        errD_fake = criterion(output, label)
        errD_fake.backward()
        D_G_z1 = output.mean().item()

        # Compute total D loss
        errD = errD_real + errD_fake
        optimizerD.step()

        ############################
        # (2) Update G network
        ###########################
        netG.zero_grad()
        label.fill_(real_label)  # Fake labels are real for generator cost

        # Since we just updated D, perform a forward pass of all fakes through D
        output = netD(sketches, fake_images)
        errG = criterion(output, label)
        errG.backward()
        D_G_z2 = output.mean().item()
        optimizerG.step()

        # Save Losses for plotting later
        G_losses.append(errG.item())
        D_losses.append(errD.item())

        iters += 1

        # Output training stats
        if i % 100 == 0:
            print(f'[{epoch+1}/{num_epochs}][{i}/{len(dataloader)}]\t'
                  f'Loss_D: {errD.item():.4f}\tLoss_G: {errG.item():.4f}\t'
                  f'D(x): {D_x:.4f}\tD(G(z)): {D_G_z1:.4f} / {D_G_z2:.4f}')

    # Save generated images at certain epochs
    if (epoch + 1) % 10 == 0 or epoch == 0:
        with torch.no_grad():
            fake = netG(sketches).detach().cpu()
        img_list.append(fake)
        # Create directory to save generated images
        os.makedirs('output', exist_ok=True)
        # Save images to disk
        torchvision.utils.save_image(fake, f'output/fake_samples_epoch_{epoch+1}.png', normalize=True)

print("Training Finished!")

# ================================
# 7. Visualization of Training Progress
# ================================

# Plot the losses
plt.figure(figsize=(10,5))
plt.title("Generator and Discriminator Loss During Training")
plt.plot(G_losses, label="G", alpha=0.7)
plt.plot(D_losses, label="D", alpha=0.7)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Display some generated images
# Grab a batch of real images from the last batch
real_batch = real_images[:8].detach().cpu()

# Grab a batch of fake images from the last batch
fake_images = fake[:8].detach().cpu()

# Plot real and fake images
plt.figure(figsize=(15,4))
for i in range(8):
    # Real Images
    plt.subplot(2,8,i+1)
    plt.axis("off")
    plt.title("Real")
    plt.imshow(np.transpose(real_batch[i], (1,2,0)) * 0.5 + 0.5)  # Denormalize

    # Fake Images
    plt.subplot(2,8,8+i+1)
    plt.axis("off")
    plt.title("Fake")
    plt.imshow(np.transpose(fake_images[i], (1,2,0)) * 0.5 + 0.5)  # Denormalize
plt.show()

# ================================
# 8. Generating Images from New Sketches
# ================================

def generate_image(generator, sketch_path, transform, device):
    """
    Generates a realistic face image from a sketch.

    Parameters:
    - generator: Trained Generator model.
    - sketch_path: Path to the sketch image.
    - transform: Transformations to apply to the sketch.
    - device: Device to perform computations on.

    Returns:
    - Generated image tensor.
    """
    generator.eval()
    sketch = Image.open(sketch_path).convert('RGB')
    sketch = transform(sketch).unsqueeze(0).to(device)
    with torch.no_grad():
        fake = generator(sketch).detach().cpu()
    return fake

# Example usage
# Select a sample sketch from the training set
sample_sketch_filename = os.listdir(sketch_dir)[0]
sample_sketch_path = os.path.join(sketch_dir, sample_sketch_filename)
print(f"Sample Sketch: {sample_sketch_filename}")

# Generate image
generated_image = generate_image(netG, sample_sketch_path, transform, device)

# Plot the generated image alongside the input sketch
plt.figure(figsize=(10,5))

# Input Sketch
sketch = Image.open(sample_sketch_path).convert('RGB')
sketch = transform(sketch)
sketch_np = sketch.cpu().numpy()
sketch_np = np.transpose(sketch_np, (1,2,0)) * 0.5 + 0.5  # Denormalize

plt.subplot(1,2,1)
plt.axis("off")
plt.title("Input Sketch")
plt.imshow(sketch_np)

# Generated Image
fake_img = generated_image[0]
fake_img_np = np.transpose(fake_img, (1,2,0)) * 0.5 + 0.5  # Denormalize

plt.subplot(1,2,2)
plt.axis("off")
plt.title("Generated Image")
plt.imshow(fake_img_np)

plt.show()

# ================================
# 9. Saving the Models
# ================================

# Create a directory to save models
os.makedirs('saved_models', exist_ok=True)

# Save the generator and discriminator models
torch.save(netG.state_dict(), 'saved_models/cgan_generator.pth')
torch.save(netD.state_dict(), 'saved_models/cgan_discriminator.pth')

print("Models saved to 'saved_models/' directory.")

# ================================
# 10. Loading the Models (Optional)
# ================================

# To load the models later, use the following code:

# Initialize models
# loaded_netG = Generator(input_channels=num_channels, output_channels=num_channels, feature_maps=64).to(device)
# loaded_netD = Discriminator(input_channels=num_channels, output_channels=num_channels, feature_maps=64).to(device)

# Load state dictionaries
# loaded_netG.load_state_dict(torch.load('saved_models/cgan_generator.pth'))
# loaded_netD.load_state_dict(torch.load('saved_models/cgan_discriminator.pth'))

# Set models to evaluation mode
# loaded_netG.eval()
# loaded_netD.eval()

# print("Models loaded and set to evaluation mode.")

# ================================
# 11. Conclusion
# ================================

print("""
Conditional GAN (cGAN) Training Complete!

You have successfully implemented a Conditional GAN to generate realistic face images from sketches using the Person Face Sketches dataset.

Key Steps Covered:
1. Environment Setup
2. Accessing and Preparing the Dataset
3. Defining the cGAN Architecture
4. Training the cGAN
5. Generating Images from New Sketches
6. Saving and Loading Models
7. Visualizing Results

Next Steps:
- Experiment with different architectures for the Generator and Discriminator to improve image quality.
- Implement advanced loss functions like L1 or perceptual loss to enhance generated images.
- Use more sophisticated evaluation metrics such as Inception Score (IS) or Frechet Inception Distance (FID).
- Explore data augmentation techniques to make the model more robust.

Happy Coding!
""")
```
